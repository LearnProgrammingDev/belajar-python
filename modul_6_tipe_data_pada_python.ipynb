# Tipe data python
tipe data berfungsi untuk menampung angka atau string text pada python dan python merupakan salah satu bahasa yang memiliki tipe data yang unik berikut tabelnya tipe data python pada gambar di bawah:


tipe-data-python.png

```python
#tipe data boolean
print(True)
print(False)

#tipe data string
print("belajar python menyenangkan") # menggunakan kutip ""
print('belajar python menyenangkan') #  mengunakan kutip ''
print("""belajar python menyenangkan""") # menggunakan kutip """

#tipe data integer
print(10)
print(1000000)

#tipe data float / bilangan berkoma
print(3.14)
print(.14)

#tipe data hexadecimal
print(0xA9)

#tipe data octal
print(0o34)

#tipe data complex
print(complex(2,3))
print(5j)

#tipe data list
print([1,2,3,4,5])
print(["satu","dua","tiga"])
print(["satu",1,2.3])
print(["senin", "selasa", "rabu", "kamis", "jum'at", "sabtu", "minggu"])

#tipe data tuple
print((1,2,3))
print(("satu","dua","tiga"))

#tipe data dictionary
print({"nama":"budi","umur":22})
print({1:"satu",2:"dua"})
print({True:"benar",False:"salah"})
```
